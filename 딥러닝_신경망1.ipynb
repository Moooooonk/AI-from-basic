```python
#기초 퍼셉트론

def AND(x1, x2): # 입력 신호
    w1, w2, theta = 0.5, 0.5, 0.7 #가중치 및 임계값 설정
    tmp = x1*w1 + x2*w2

    if tmp >= theta:
        return 0
    elif tmp < theta:
        return 1

    print(theta)
```

***퍼셉트론***

1. X(입력)에서 W(가중치)를 곱한 후 모두 더한 수 비선형 활성화 함수를 통과시킴. 통과시키면 어떻게 되는지, 들어오는 수치가 무엇을 의미하는지 정확히 모름
Bias(편향, 1)을 이때 더하는데, 이유는 뉴런이 활성화함수 x축을 따라 수평으로 이동할 수 있게 함. 바이어스가 어떻게 수평이동에 도움을 주는지 모르겠음.
  
-> 값이 뉴런의 출력값이 되려면 비선형성이 필요해. 그래서 활성화 함수를 통과시켜서 비선형성을 추가한다.

바이어스의 역할: 바이어스는 선형 방정식의 절편 역할을 한다. 바이어스를 더하면, 활성화 함수의 그래프가 x축을 따라 평행 이동할 수 있음.


2. 시그모이드 함수는 0과 1 사이로 들어오는 값을 압축해주고, 미분 가능한 것이 장점. 미분이 가능하면 역전파 가능.


3. 활성화 함수의 정확한 의미와 미분이 왜 가능해야 하는지 등등 잘 모르겠음.

->활성화 함수는 비선형성을 추가하는 역할을 하는데,  활성화 함수가 없다면, 신경망은 단순한 선형 변환을 계속하게 되고, 결국 복잡한 패턴을 학습하지 못한다.

미분이 가능해야 하는 이유: 신경망은 역전파(Backpropagation) 알고리즘을 사용해서 학습하는데, 역전파 과정에서 미분(gradient) 이 필요하다.
미분을 통해 오차(손실) 를 줄이는 방향으로 가중치를 업데이트할 수 있다.


4. 우리가 원하는 것은 신경망이 비선형적 데이터를 처리할 수 있게 하는 것


5. 비선형적인 함수가 신경망의 크기를 줄이는데, 어떻게?

-> 비선형성을 추가하면, 적은 층과 뉴런으로도 더 강력한 표현력을 가질 수 있음.
예를 들어, XOR 문제를 풀려면 선형 모델에서는 두 개의 층 이상이 필요하지만, 비선형 활성화 함수를 사용하면 한 층에서도 가능함.
즉, 비선형성이 있으면 더 적은 뉴런과 층으로 복잡한 문제를 해결할 수 있다.


7. 가중치는 어떻게 설정되며 변화하는지도 정확히 모르겠음.

-> 가중치는 처음에 무작위 값(random initialization) 으로 설정되고, 학습이 진행되면서 역전파(Backpropagation) + 경사 하강법(Gradient Descent) 을 통해 가중치가 조정됨.
목표는 손실 함수(Loss Function)를 최소화하는 방향으로 가중치를 업데이트하는 것.


8. 모든 뉴런에 바이어스가 있는 것이 맞나?

-> 일반적으로 모든 뉴런에는 바이어스가 존재함.
바이어스가 없으면 뉴런이 원점을 기준으로만 학습하게 되고, 데이터에 대한 표현력이 떨어짐.
하지만 특정한 경우(예: 정규화 기법을 사용할 때) 바이어스를 제거할 수도 있음.

```python
#배열 활용 퍼셉트론

# 퍼셉트론의 AND, NAND, OR은 가중치와 편향의 차이. 나머지는 일치
import numpy as np

def AND(x1, x2):

    x = np.array([x1, x2]) # 입력
    w = np.array([0.5, 0.5]) # 가중치
    b = -0.7 # 편향 (연산 후 편향보다 높은 값일 때 1 반환)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x*w)+b

    if tmp <= 0:
        return 0
    else:
        return 1

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(x*w) + b

    if tmp <= 0:
        return 0
    else:
        return 1

#다층 퍼셉트론(XOR)

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

#각 함수의 역할 및 특성

1. Step Function (계단 함수)
역할 및 특성:
입력 값이 0보다 크면 1, 작으면 0을 반환하는 이진 활성화 함수.
비선형성을 도입하지만, 미분이 불가능하여 신경망 학습(역전파)에 적합하지 않음.
특정 임계값을 기준으로만 작동하여 세밀한 변화 반영이 어려움.
주로 퍼셉트론 모델에서 사용되었으나, 현대 신경망에서는 거의 사용되지 않음.

2. Sigmoid Function (시그모이드 함수)
역할 및 특성:
출력 범위가 (0,1) 사이로 제한되며, 확률적 해석이 가능하여 분류 문제에서 사용됨.
작은 입력 값에서는 0에 가깝고, 큰 입력 값에서는 1에 가까워지는 S자 곡선을 가짐.
하지만 기울기 소실 문제(vanishing gradient problem) 발생 가능:
극단적인 입력에서는 미분값이 0에 가까워져서 학습이 어려워짐.
연산량이 많아 최근에는 주로 출력층에서 확률 해석을 위해 사용됨.

3. ReLU (Rectified Linear Unit, 렐루)
역할 및 특성:
0 이하의 입력 값은 0으로 변환하고, 양수는 그대로 유지.
음수 값을 완전히 제거하므로 희소성(Sparsity) 을 유도하고, 기울기 소실 문제를 완화함.
하지만 음수 입력에 대해 미분값이 0이 되는 죽은 뉴런 문제(Dead Neuron Problem) 가 발생할 수 있음.

4. Leaky ReLU (누수 렐루)
역할 및 특성:
ReLU의 단점을 보완하기 위해 음수 값 영역에서도 작은 기울기를 유지함.
ReLU와 유사하지만 죽은 뉴런 문제를 방지할 수 있음.
leaky_relu(x) = max(0.1x, x) 에서 0.1은 작은 기울기 계수이며, 다른 값으로 조정할 수도 있음.

5. Tanh (하이퍼볼릭 탄젠트 함수)
역할 및 특성:
Sigmoid와 유사하지만 출력 범위가 (-1,1)으로 확장됨.
0을 중심으로 대칭이므로, 평균이 0에 가깝게 유지되어 학습에 유리함.
하지만 여전히 극단적인 입력에서 기울기 소실 문제가 발생할 수 있음.

6. Softmax Function (소프트맥스 함수)
역할 및 특성:
다중 클래스 분류 문제에서 사용되며, 입력 벡터를 확률 분포로 변환.
각 출력이 전체 입력 값의 영향을 받으며, 모든 출력의 합이 1이 됨.
개별 뉴런이 아니라 출력층 전체에 적용되는 함수.

7. MoonLu Function (문루 함수)
역할 및 특성:
ReLU와 유사하지만, 음수 입력에서 단순 선형 변환이 아니라 지수적 증가를 허용.
ReLU보다 부드럽고, 작은 음수 입력에 대한 정보 손실을 줄일 수 있음.
미분값이 0이 되는 영역이 없으므로 죽은 뉴런 문제를 방지할 가능성이 있음.

```python
#3층 퍼셉트론 구현

import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

def step_function(x):
    return np.array(x > 0, dtype=np.int32)

def tanh(x):
  return np.tanh(x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.maximum(0.1*x, x)

def softmax(a): #소프트맥스 함수는 지수함수로 인한 거대수 문제 발생 가능. 이때, 나눗셈 계산시 수치 불안정
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y

def moonlu(x):
    return np.where(x < 0, np.exp(x), x + 1)

def identify_function(x):
    return x

def init_network(): #가중치, 편향 초기화
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])

    return network

def forward(network, x, activation_hidden, activation_output):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = activation_hidden(a1)  # 은닉층 활성화 함수 적용
    a2 = np.dot(z1, W2) + b2
    z2 = activation_hidden(a2)  # 은닉층 활성화 함수 적용
    a3 = np.dot(z2, W3) + b3
    y = activation_output(a3)   # 출력층 활성화 함수 적용

    return y

# 여러 개의 입력 샘플을 포함하는 2D 배열
x = np.array([
    [1.0, 0.5],  # 첫 번째 입력
    [0.2, 0.8],  # 두 번째 입력
    [0.9, 0.1]   # 세 번째 입력
])
# 활성화 함수 실험 및 비교
activations = {
    "Step": step_function,
    # Step Hidden Layer Output:
    # [[0.4 0.9]
    #  [0.4 0.9]
    #  [0.4 0.9]]
    # -> Step 함수는 0 또는 1의 이진 값만 출력하므로 비선형 변환이 극단적임.
    # -> 미분이 불가능하고, 작은 변화에 대한 반응이 없음.

    "Sigmoid": sigmoid,
    # Sigmoid Hidden Layer Output:
    # [[0.31682708 0.69627909]
    #  [0.31489781 0.69197608]
    #  [0.31381012 0.68957964]]
    # -> 0~1 사이의 값으로 변환하여 확률적 해석 가능.
    # -> 작은 변화에 민감하지만, 큰 입력값에서 미분값이 매우 작아지는 (기울기 소실) 문제가 발생.

    "ReLU": relu,
    # ReLU Hidden Layer Output:
    # [[0.426  0.912 ]
    #  [0.3768 0.8044]
    #  [0.3516 0.7498]]
    # -> 0 이상의 값은 그대로 전달하고, 0 이하는 0으로 설정하여 비선형성 유지.
    # -> 기울기 소실 문제 해결, 하지만 모든 음수 입력이 0이 되어 뉴런이 죽는 문제 발생 가능.

    "Leaky ReLU": leaky_relu,
    # Leaky ReLU Hidden Layer Output:
    # [[0.426  0.912 ]
    #  [0.3768 0.8044]
    #  [0.3516 0.7498]]
    # -> ReLU와 비슷하지만, 음수 입력에 대해 작은 기울기(0.1)를 부여하여 뉴런이 완전히 죽는 문제를 방지.

    "MoonLU" : moonlu,
    # MoonLU Hidden Layer Output:
    # [[1.086  2.392 ]
    #  [1.0368 2.2844]
    #  [1.0116 2.2298]]
    # -> 음수 값에서는 지수함수 적용, 양수 값에서는 선형적으로 증가.
    # -> 모든 점에서 미분 가능하고, 음수 입력값에서 뉴런이 죽는 문제 방지.

    "Tanh": tanh,
    # Tanh Hidden Layer Output:
    # [[0.30546241 0.65635873]
    #  [0.29351665 0.62870805]
    #  [0.284452   0.60822257]]
    # -> -1 ~ 1 범위로 변환하여 평균이 0에 가까워지는 장점이 있음.
    # -> Sigmoid와 유사하지만, 기울기 소실 문제가 여전히 존재.

    "Softmax": softmax
    # Softmax Hidden Layer Output:
    # [[0.15265201 0.32046593]
    #  [0.15140928 0.31778048]
    #  [0.15092482 0.31673971]]
    # -> 출력을 확률 분포로 변환하여 다중 클래스 분류 문제에서 사용.
    # -> 출력값의 총합이 1이 되며, 가장 큰 값에 상대적으로 높은 확률을 부여.
}
# 결과 저장
network = init_network()
results = {name: forward(network, x, func, softmax) for name, func in activations.items()}

# 그래프를 입력별로 분리하여 시각화
fig, axes = plt.subplots(3, len(activations), figsize=(20, 12))

input_labels = ["Input 1", "Input 2", "Input 3"]

for i, (name, output) in enumerate(results.items()):
    for j in range(3):  # 3개의 입력값에 대해 각각 그리기
        axes[j, i].bar(range(output.shape[1]), output[j], alpha=0.8, color=['b', 'g'])
        axes[j, i].set_title(f"{name} ({input_labels[j]})")
        axes[j, i].set_ylim(0, 1.2)  # 값의 범위를 일정하게 설정

plt.tight_layout()
plt.show()
for name, activation in activations.items():
    y = forward(network, x, activation, identify_function)
    print(f"{name} Hidden Layer Output: {y}")

# 최종 출력층에 Softmax 적용
y_softmax = forward(network, x, sigmoid, softmax)
print(f"Softmax Output: {y_softmax}") #Softmax Output: [0.40625907 0.59374093]

# 그래프 그리기
plt.figure(figsize=(10, 6))
for name, act in activations.items():
    plt.plot(x, act(x), label=name)

plt.title("Activation Functions Output")
plt.xlabel("Input")
plt.ylabel("Output")
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()

# 2D 입력값 생성 (그리드 형태)
x1 = np.linspace(-2, 2, 50)
x2 = np.linspace(-2, 2, 50)
X1, X2 = np.meshgrid(x1, x2)
X_input = np.c_[X1.ravel(), X2.ravel()]  # (2500, 2) 형태로 변환

for name, activation in activations.items():
    Z = activation(X_input[:, 0] + X_input[:, 1]).reshape(X1.shape)

    # 2D 히트맵 플롯
    plt.figure(figsize=(8, 6))
    plt.contourf(X1, X2, Z, levels=20, cmap='coolwarm')
    plt.colorbar(label="Activation Output")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f"{name} Activation Function Heatmap")
    plt.show()

    # 3D 서피스 플롯
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X1, X2, Z, cmap='coolwarm')

    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_zlabel("Activation Output")
    ax.set_title(f"{name} Activation Function 3D Plot")

    plt.show()

# 개별 그래프 그리기
x_vals = np.linspace(-2, 2, 100)  # 입력 값 범위 설정

plt.figure(figsize=(12, 12))

for i, (name, act) in enumerate(activations.items(), 1):
    plt.subplot(3, 3, i)  # 3x3 서브플롯으로 배치
    plt.plot(x_vals, act(x_vals), label=name, color='b')
    plt.title(name)
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()

a = np.array([1010, 1000, 990])
s1 = np.exp(a) / np.sum(np.exp(a))
print(s1) #[nan nan nan], 계산 불가

c = np.max(a) # c = 1010 (배열 중 최대값)
s2 = np.exp(a - c) / np.sum(np.exp(a - c))
print(s2) # [9.99954600e-01 4.53978686e-05 2.06106005e-09], 이처럼 입력 신호 중 최대값을 빼주면 계산 가능

```

```python
import numpy as np

def softmax(x):
    """Softmax 함수 구현"""
    e_x = np.exp(x - np.max(x))  # 오버플로우 방지를 위해 최댓값으로 빼줌
    return e_x / e_x.sum(axis=0)

# 입력 데이터 (각 열은 샘플, 각 행은 특성)
X = np.array([[0.5, 2.0, 1.5],
              [1.0, 2.5, 3.0],
              [2.0, 1.0, 0.5]])

# 가중치 행렬 (각 행은 클래스, 각 열은 특성)
W = np.array([[0.2, 0.5, 0.1],
              [0.8, 0.3, 0.6],
              [0.4, 0.7, 0.9]])

# 편향 벡터 (각 요소는 클래스)
b = np.array([0.1, 0.2, 0.3])

# 각 샘플에 대한 점수 계산
scores = np.dot(W, X) + b.reshape(-1, 1)

# Softmax 함수 적용
probabilities = softmax(scores)

# 결과 출력
print("입력 데이터:")
print(X)
print("\n가중치 행렬:")
print(W)
print("\n편향 벡터:")
print(b)
print("\n점수:")
print(scores)
print("\n확률:")
print(probabilities)
```
